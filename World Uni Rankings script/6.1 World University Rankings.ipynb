```python
# Importing libraries
import pandas as pd
import numpy as np
import os
```

```python
# Importing the dataframe
df_rankings = pd.read_csv(r'C:\Users\49157\Downloads\Data analysis course\Advanced analytics and dashboard designs\World University Rankings\cwurData.csv', index_col = False)
```

```python
# Data consistency checks
df_rankings.describe()
```

```python
# Checking for mixed data types
for col in df_rankings.columns.tolist():
  weird = (df_rankings[[col]].applymap(type) != df_rankings[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_rankings[weird]) > 0:
    print (col)
```

```python
# Finding missing values
df_rankings.isnull().sum()
```

```python
# To view the missing values in the broad impact column
df_nan = df_rankings[df_rankings['broad_impact'].isnull() == True]
df_nan
```

```python
# Checking the rows and columns again 
df_rankings.shape
```

```python
# Addressing the missing values in the broad impact column by replacing them with the mean values 
mean_broad_impact = df_rankings.groupby('institution')['broad_impact'].transform('mean')
df_rankings['broad_impact'] = df_rankings['broad_impact'].fillna(mean_broad_impact)
```

```python
# Checking the missing values again to ensure the mean values has been replaced
df_rankings.isnull().sum()
```

```python
# Checking the value in the broad impact column with one entry
df_rankings[df_rankings['broad_impact'].isnull() == True]
```

```python
# I dropped the williams college institute because it has only one entry
df_rankings = df_rankings[df_rankings.institution != 'Williams College']
```

```python
# Checking if the broad impact column still shows missing values
df_rankings.isnull().sum()
```

```python
# Checking the rows and columns again 
df_rankings.shape
```

```python
# Finding duplicates
df_dups = df_rankings[df_rankings.duplicated()]
df_dups
```
